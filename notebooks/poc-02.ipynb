```python
import pandas as pd
import numpy as np
from collections import Counter
import glob
import os
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
import networkx as nx
from pyvis.network import Network
```

```python
df = {}
for file in glob.glob("../data/raw/scenario01/csv/*.csv"):
    tmp_df = pd.read_csv(file)
    file_sep = file.split(os.sep)
    df_name = file_sep[len(file_sep) - 1].split('.')[0]
    df[df_name] = tmp_df
```

```python
encounters = df['encounters']
encounters['row_number'] = encounters.sort_values(['START'], ascending=[False]) \
             .groupby(['PATIENT']) \
             .cumcount() + 1
last_encounters = encounters[encounters['row_number'] == 1]
```

```python
patients = df['patients']
patients
```

```python
patients_last_encounter = patients.merge(last_encounters, left_on='Id', \
                                         right_on='PATIENT', suffixes=('_patient', '_encounter'))
```

```python
patients_last_encounter['death_threshold'] = pd.to_datetime(patients_last_encounter['DEATHDATE']).dt.tz_localize(None) - \
pd.to_datetime(patients_last_encounter['START']).dt.tz_localize(None) <= datetime.timedelta(days=7)
```

```python
patients_last_encounter
```

```python
patients_encouterclasses = encounters.groupby(['PATIENT', 'ENCOUNTERCLASS']).size().unstack('ENCOUNTERCLASS', fill_value=0).reset_index()
```

```python
patients_encouterclasses
```

```python
patients_last_encounter = patients_last_encounter.merge(patients_encouterclasses, left_on='Id_patient', \
                              right_on='PATIENT', suffixes=('_patient', '_encounter_classes'))
```

```python
patients_last_encounter.columns
```

```python
conditions = df['conditions']
```

```python
conditions['row_number'] = conditions.sort_values(['STOP', 'START'], ascending=[False, False]) \
             .groupby(['PATIENT']) \
             .cumcount() + 1
```

```python
last_conditions = conditions[conditions['row_number'] == 1]
```

```python
len(list(set(last_conditions['PATIENT'])))
```

```python
last_conditions
```

```python
patient_last_conditions = last_conditions.groupby(['PATIENT', 'CODE']).size(). \
unstack('CODE', fill_value=0).reset_index()
```

```python
patient_last_conditions
```

```python
patients_last_encounter_conditions = patients_last_encounter.merge(patient_last_conditions, left_on='Id_patient', \
                              right_on='PATIENT', suffixes=('_patient_encounter_classes', '_conditions'))
```

```python
patients_last_encounter_conditions.columns
```

```python
def get_days_alive(row):
    try:
        return pd.Timedelta(
            pd.to_datetime(
                row['DEATHDATE'], errors='coerce'
            ).tz_localize(None) - \
            pd.to_datetime(
                row['BIRTHDATE'], errors='coerce'
            ).tz_localize(None)
        ).days
    except:
        return pd.NaN
```

```python
patients_last_encounter_conditions['days_alive'] = patients_last_encounter_conditions.apply(get_days_alive, axis=1)
```
